```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import FuncFormatter

data = pd.read_csv("/content/drive/MyDrive/SAAS/jobs_salaries_data.csv", encoding='unicode_escape')

# Filter data for USD currency and United States residence
datafil1 = data[data['salary_currency'].isin(["USD"]) & data['employee_residence'].isin(["United States"])].head(100)

# Plotting the filtered data
x = datafil1['job_title']
y = datafil1['salary']

# Fungsi untuk menyingkat gaji menjadi format "1k"
def salary_formatter(x, pos):
    return f"{x/1000:.0f}k"

plt.bar(x, y, align='center')

plt.title('Job Titles and Salaries for USD Currency in the United States')
plt.xlabel('Job Title')
plt.ylabel('Salary (USD)')

# Mengatur rotasi label sumbu x
plt.xticks(rotation=45, ha='right')

# Mengatur formatter untuk sumbu y
plt.gca().yaxis.set_major_formatter(FuncFormatter(salary_formatter))

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("/content/drive/MyDrive/SAAS/jobs_salaries_data.csv", encoding='unicode_escape')

# Filter data for USD currency and United States residence
datafil1 = data[data['salary_currency'].isin(["USD"]) & data['employee_residence'].isin(["United States"])].head(100)

# Group data by job title and find the maximum salary for each title
max_salaries = datafil1.groupby('job_title')['salary'].max()

# Plotting the filtered data
x = max_salaries.index
y = max_salaries.values

# Create bar chart with adjusted bar_width
bar_width = 0.6
bars = plt.bar(x, y, align='center', width=bar_width, color="lightblue")

plt.title('Job Titles and Highest Salaries for USD Currency in the United States')
plt.xlabel('Job Title')
plt.ylabel('Highest Salary (USD)')

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right')

# Add labels above each bar with the highest salary value (with decreased font size)
for bar, job, salary in zip(bars, x, y):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05, f"${salary:,.0f}", ha='center', fontsize=8)

# Adjust layout to prevent overlapping
plt.tight_layout()

plt.show()
```
